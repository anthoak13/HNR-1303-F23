# Course Learning Objectives

1. I can study a problem (i.e. question) in physics, biology, economics, or social sciences; apply simplifying assumptions; and develop a mathematical model (equations) to describe the system: Saw model to fit and understand trend.

1.  I can apply the appropriate mathematical technique to solve the equations. (This includes selected topics in calculus, vector algebra, linear algebra, numerical methods and statistics.). Note vector algera (all the models/functions)

1.  I can graph model data, interpret the graph, and use the graph to make predictions: Yes whole unit

1.  I can compare measured data to model data and can judge the accuracy of the model: Yes, most things

1.  I can refine a model based on a comparison with measured data or unreasonable predictions: Yes (Covid, exponential to logarithm)

1.  When necessary, I can write a computer program in Python to graph data or perform calculations: Yes, entire unit

1.  I can use graphs, equations, and prose to write or present a data- and code-driven narrative to explain a model, predictions, and validation: Yes, Beer's law is example and Barbie stocks

1.  I can work in a group on a project to solve a problem using mathematical modeling and can demonstrate all aspects of modeling including construction, analysis, predictions, and refinement: Yes (asked to explain things in group and present/group assesment).

1.  I can describe ethical implications of decisions made in the modeling process and how they affect conclusions and predictions made with the model: Mostly around predictions and limitations of data.

1.  I can work productively in a group, constructively reflect on a working group dynamic, and develop/implement strategies to improve productivity in a group: Yes, working in goups most of the time.
