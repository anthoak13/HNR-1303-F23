## Exercise 1

The radioactive decay of atoms is used in treatment of cancer. In particular, the [decay of Strontium-90 atoms is used to treat bone tumors](https://www.cdc.gov/nceh/radiation/emergencies/isotopes/strontium.htm). Strontium-90 atoms decay through emission of electrons. This particular decay process is called beta decay. 

A single strontium-90 atom decays into a electron, an antineutrino, and a yttrium-90 atom. If $Q$ represents the number of strontium-90 atoms remaining (undecayed) in a sample, the rate of change of $Q$ is

$$rate= \frac{\Delta Q}{\Delta t} = -r Q$$

where the decay rate constant is $r=0.0241$ in units of $1/year$.

Compute and graph $Q$ as a function of time up to 20 years, using a time step of 0.01 year and an initial number of 1000 strontium-90 atoms. What is the number of strontium-90 atoms, $Q$, at $t=10$ years?

````python
## define constants
r = 0.0241 # 1/year
dt = 0.01 # years

# define variables and their initial values
Q = 1000
t = 0

# create empty lists for storing data
tdata = []
Qdata = []

# append initial values of t and P to our lists
tdata.append(t)
Qdata.append(Q)

# loop
while t<20:
    rate = -r*Q
    Q = Q + rate*dt
    t = t + dt
    
    tdata.append(t)
    Qdata.append(Q)

print(Q)
plt.figure(figsize=(8,6))
plt.title('Atoms vs time')
plt.xlabel('time (years)')
plt.ylabel('Number of atoms')
plt.plot(tdata,Qdata,'b-')
plt.grid()
plt.show()
````

At 10 years there are about 785 atoms of $\text{Sr}^{90}$ left.

$$rate = \frac{\Delta T}{\Delta t} = -k(T-20)$$

where $k=0.1$ in units of $1/minute$ and $T$ is in degrees celsius.

(Use a unit for time in minutes, and convert your time step to minutes, knowing that there are 60 s in one minute.)

````python
## define constants
k = 0.1 # rate in units 1/minutes
dt = 1/60 # minutes

# define variables and their initial values
T = 80 # Coffee temp (C)
t = 0

# Empty lists to use for plotting
tdata = []
Tdata = []

# loop while there is still a lone to pay
while t < 60:
    rate = -k*(T-20) 
    
    T = T + rate*dt

    t = t + dt
    tdata.append(t)
    Tdata.append(T)

    # Add code to print temperature at different times this could also be read off the graph
    if round(t,2) == 5:
        print(f"At 5 minutes, the temperature of the coffee is {T} deg C")
    if round(T,1) == 40:
        print(f"At {t} minutes, the temperature of the coffee is {T} deg C")
        
plt.plot(tdata,Tdata)
plt.title("Temperature of coffee over time")
plt.xlabel("Time(min)")
plt.ylabel("Temperature (C)")
plt.show()
````
